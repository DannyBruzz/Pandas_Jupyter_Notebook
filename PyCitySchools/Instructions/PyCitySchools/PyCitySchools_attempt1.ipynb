```python
import pandas as pd
import os
import csv

school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"


school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#Local Government Area Summary
lga_total_schools = school_data_complete["school_name"].nunique()
lga_total_students = school_data_complete["Student ID"].nunique()
lga_total_budget = school_data["budget"].sum()
lga_ave_maths = school_data_complete["maths_score"].mean()
lga_ave_reading = school_data_complete["reading_score"].mean()
lga_pass_math = len(school_data_complete.loc[school_data_complete["maths_score"] >=50, :])
lga_ave_passmath = lga_pass_math / lga_total_students * 100
lga_pass_read = len(school_data_complete.loc[school_data_complete["reading_score"] >=50, :])
lga_ave_passread = lga_pass_read / lga_total_students * 100
lga_pass_both = len(school_data_complete.loc[(school_data_complete["reading_score"] >=50) & (school_data_complete["maths_score"] >=50), :])
lga_pass_b = lga_pass_both / lga_total_students * 100
lga_summary_df = pd.DataFrame({"Total Schools": lga_total_schools, 
                               "Total Students": lga_total_students, 
                               "Total Budget": lga_total_budget, 
                               "Average Maths Score": lga_ave_maths, 
                               "Average Reading Score": lga_ave_reading, 
                               "% Passing Maths": lga_ave_passmath, 
                               "% Passing Reading": lga_ave_passread, 
                               "% Overall Passing": lga_pass_b}, index=[0])
lga_summary_df.head()
```

```python
#School Summary
sch = school_data_complete.groupby(["school_name"])
sch_type = sch["type"].unique()
sch_size = sch["size"].unique()
sch_budget = sch["budget"].unique()
sch_b_p_stu = sch_budget / sch_size
sch_math_ave = sch["maths_score"].mean()
sch_read_ave = sch["reading_score"].mean()

maths_filter = school_data_complete.loc[school_data_complete["maths_score"] >=50, :]
pass_math_count = maths_filter.groupby(["school_name"]).count()["maths_score"] 
sch_pass_maths_ave = pass_math_count / sch_size
sch_pass_maths_ave = sch_pass_maths_ave.astype(float)

reading_filter = school_data_complete.loc[school_data_complete["reading_score"] >=50, :]
pass_reading_count = reading_filter.groupby(["school_name"]).count()["reading_score"]
sch_pass_reading_ave = pass_reading_count / sch_size
sch_pass_reading_ave = sch_pass_reading_ave.astype(float)

passed_filter = school_data_complete.loc[(school_data_complete["reading_score"] >=50) & (school_data_complete["maths_score"] >=50), :]
pass_both_count = passed_filter.groupby(["school_name"]).count()["School ID"]
sch_pass_ave = pass_both_count / sch_size
sch_pass_ave = sch_pass_ave.astype(float)
sch_pass_ave

sch_summary_df = pd.DataFrame({"School Type": sch_type, 
                               "Total Students": sch_size,
                               "Total School Budget": sch_budget,
                               "Per Student Budget": sch_b_p_stu,
                               "Average Maths Score": sch_math_ave,
                               "Average Reading Score": sch_read_ave,
                               "% Passing Maths": sch_pass_maths_ave,
                               "% Passing Reading": sch_pass_reading_ave,
                               "% Overall Passing": sch_pass_ave})
                              
sch_summary_df.head(15)

```

```python
#Top Performing Schools (By % Overall Passing)
top_schools = sch_summary_df.sort_values("% Overall Passing", ascending=False)
top_schools.head(5)
```

```python
#Bottom Performing Schools (By % Overall Passing)
bottom_schools = sch_summary_df.sort_values("% Overall Passing")
bottom_schools.head(5)
```
